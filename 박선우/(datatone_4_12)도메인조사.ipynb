# 원화 가치 변동의 원인(일반적)

<기본적으로 수요가 많은데 공급이 적으면 가치가 상승>

► "환율 상승 = 원화 가치 하락" 원인

1. 미국의 고금리 정책 유지 ➞ 미국이 한국보다 훨씬 안정적인 시장

      
      ex) 미국의 금리 3% vs 우리나라는 2% : 당연히 미국에 투자 ⇒ 한국 자본 유출(한국투자한 외국거까지) ⇒ 원화가치하락
         (한미 금리 차이가 1.5%p 이상 벌어지면 자본 유출 우려 → 원화 가치 하락(?))

         미국주식투자의 경우 , 5000원 주고 삼 → 주가는 같은데 7500원 됨

2. 중국 경기 둔화

      
      중국은 한국의 최대 수출국 하지만 소비/투자 둔화 ➞ 한국 수출 감소 ➞ 무역수지 악화(적자) → 원화 약세 압력(더 싸게 팔야함)

3. 반도체 수출 부진 및 무역수지 적자


      반도체 회복세가 약화 & 유가 상승 등 ➞ 무역수지가 적자 전환 ➞ (수출부진하면 우리나라의 경쟁력 약화 그럼 더 싸게 팔아야 함)약세
      특히 원유 수입국인 한국은 유가 상승 시 환율 상승 영향 큼


4. 지정학적(지리적) 리스크


      북한 리스크, 중동/우크라이나 전쟁 등 불확실성 높아지면 안전자산인 달러 수요 증가
      상대적으로 원화는 위험자산 취급 (ex. 북한 공격등으로 불안정) → 원화 가치 하락



최근 원화 약세의 핵심은 미국의 고금리 유지 + 한국 경기 둔화 + 수출 부진이 복합적으로 작용

|-----------------|-------------------------|-----------------------------|
| 금리 격차       | 미국/한국 기준금리       | `722Y001` (금리)            |
| 수출입/무역수지 | 수출입 금액, 무역수지    | `901Y001`, `301Y002`       |
| 환율            | 원/달러 환율            | `731Y003`, `731Y006`       |
| 중국 경제 영향  | 대중국 수출 비중         | 관세청 또는 산업부 자료     |




► "환율 하락 = 원화 가치 상승" 원인

1. 한국의 금리 인상

  한국은행이 **기준금리를 인상**하면 원화 자산의 **수익률↑**<br>
  외국인 자금이 한국으로 **유입** → 원화 수요 증가 → 원화 강세 (환율 하락)


     ex) “미국보다 한국 금리가 더 높다” → 원화가 더 매력적 → 원화 강세

 1-1. 하지만 지금은 금리 인상을 못하고 있음. [기사참고](https://n.news.naver.com/article/656/0000128359?sid=101)
  
    ❓ 금리 인상 → 저축이 많아짐 → 경기 침체 가능성 ⬆︎

    지난 2월 한국은행(금통위)는 기준금리를 0.25%p 내렸다. 금리 인하를 통한 경기부양필요성이 커졌기 때문이다. ➡︎ **"결국 저축하지말고 돈써라, 어치피 저축해도 돈많이 안 줄거다."**

2. 무역수지 흑자 (수출 > 수입)

  (한번 더 생각해보기)
  한국 기업이 해외에 수출해서 달러로 벌어들임<br>
  그 달러를 원화로 환전하기 때문에 원화 수요↑ (공급은 없음) → 환율 하락<br>
  ex) 달러가 a 원화 a => (수출 후) 달러 2a 원화 a => 간단하게 보면, 원화를 2a 다 바꾸고 싶음 원화가치 오름


    삼성, 현대 등 수출 대기업이 많이 팔면 원화 강세 요인!
    + 우리나라 자체가 경제적으로 부양되고 경쟁력이 강화되면서 우리나라에 투자하려는 외국 자본들 증가 → 원화 가치 상승 & 환율 하락 (3번과 이어짐)

3. 외국인 투자 증가 (주식/채권)

  외국인이 코스피, 국채 등에 대규모 투자 시 → 원화로 자산 구매<br>
  원화 수요↑ → 원화 강세

  **comment. 여기까지 보면 수출이 관련이 많아보임.**

4. 글로벌 달러 약세

  전 세계적으로 달러 가치가 하락하면 상대적으로 원화 가치도 올라감<br>
  특히 미국이 금리 인하하거나 경기 둔화 신호 보이면 이런 흐름 생김

  
     미국의 경기 둔화 신호 → 내수경제를 활성화 시키기 위해서 금리 인하 → 미국 투자보다 다른나라(우리나라등)에 투자 → 미국 자본이 다른 곳으로 나감 → 달러가치가 하락 → 상대적으로 원화 가치 높아짐 → 환율 하락

5. 중국 경기 회복

  한국은 중국과 무역 연관도 매우 큼<br>
  중국이 회복되면 → 한국 **수출** 회복 기대감 (한국 것을 많이 살거니까)→ 원화 수요 증가


6. 지정학적(지리적) 안정

  북한 리스크 감소, 국제 외교 개선 등은 외국 자금 유입을 유도<br>
  외환시장 안정 → 원화 강세 흐름

  얘네 전쟁 안하겠네~ → 얘네한테 투자해도 되겠다 → 외국인 투자 활성화 (다 우리걸 사고 싶어해 근데 2개 밖에 없어! so, 비싸게 삼) → 원화 강세

# 환율이란?
환율 = 외국 돈 1단위를 사기 위해 필요한 원화 금액
예: 1달러 = 1,300원 → 이 수치가 올라가면 원화가치 하락, 내려가면 원화가치 상승



## 🔺 환율 상승 했을 때의 장단점

|장점 ✅	|단점 ❌|
|-----------------|-----------------------------|
|✅ 수출 기업에게 유리함 → 제품 가격 경쟁력 증가 (달러로 받는 돈이 더 많아짐)|	❌ 수입 물가 상승 [자급자족이 잘 되지 않는 우리나라 (석유나 자원등이 없음)] → 원자재·에너지·식품 가격 인상|
|✅ 무역수지 개선 가능 (수출↑, 수입↓)	|❌ 소비자 물가 상승 [ex.우리나라가 공기밥 2천원] → 인플레이션 유발|
|✅ 해외 관광객 유치에 유리	|❌ 해외여행/유학 비용 증가 [해외여행 가고 싶은데! 달러 오름]|
|✅ 외국인 투자자에게 매력적인 시장(저가매수)	|❌ 외채 상환 부담 증가 (달러로 갚아야 하니까 비용 ↑)|


## 🔻 환율 하락 (원화 강세 / 1달러당 원화 ↓)했을 때의 장단점

|장점 ✅	|단점 ❌|
|-----------------|-----------------------------|
|✅ 수입 물가 하락 → 소비자물가 안정 (에너지·식품 등 ↓)	|❌ 수출 경쟁력 약화 (원화 강세 → 가격 ↑)|
|✅ 해외여행, 유학, 해외 쇼핑 유리	|❌ 수출 중소기업 타격 가능|
|✅ 외채 상환 부담 감소	|❌ 무역수지 악화 가능성 (수입↑, 수출↓)|

## "결국 뭐가 좋을까?"

수출 기업에겐 원화 약세가 좋지만 내수 소비자에겐 원화 강세가 더 유리

한국처럼 수출의존도가 높은 국가에선 보통 **적절한 수준의 약세 환율(=1,200~1,300원대)**이 이상적

정리

 1. 지금 무슨 일?

  - 미국 vs 중국 관세 전쟁이 격화됨

  - 미국은 중국 제품에 145% 관세 폭탄 부과

  - 중국은 아직 대응 관세는 안 했지만, 위안화 가치를 낮추는 방향으로 조심스럽게 움직임

2. 달러는 왜 약해졌어?
  - **미국 물가 상승률(CPI)**이 낮게 나왔어 → 인플레이션 완화!

  - 그래서 시장은 “이제 미국이 금리 인하할지도 몰라”라고 생각 → 달러 가치 하락 (사람들이 달러를 덜 사게 됨)

3. 그런데 왜 원/달러 환율은 높게 유지돼?

- 원화는 달러보다 위안화 흐름에 더 민감하게 반응하고 있어

- 지금 위안화가 약세라서, → 원화도 덩달아 약세
  즉, 달러 약세 효과보다 위안화 약세 영향이 더 큼

4. 그 외 요인도 있어
  - 외국인 투자자들이 한국 주식 팔고 있음<br>
    → 원화로 된 주식을 팔고 → 달러로 바꿔서 빠져나감 → 환율↑

  - 위험회피 심리 증가: 미중 갈등, 세계 경제 불안<br>
    → 사람들은 더 안전하다고 여기는 자산(예: 미국 국채, 엔화)을 선호<br>
    → 원화는 상대적으로 덜 안전 → 원화 수요↓ → 환율↑

     "중국이 위안화 가치를 일부러 낮추고 있어요"

     "그래서 원화도 덩달아 약세입니다"

     "당분간 환율은 쉽게 안 떨어지고 1440~1500원 사이에서 움직일 것 같아요"
