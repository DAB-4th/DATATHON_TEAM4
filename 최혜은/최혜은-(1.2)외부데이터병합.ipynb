#### **한국 채권 수익률 불러오기**

```python
df_ext = pd.read_csv("/content/한국 10년 채권수익률.csv")
df_ext.head()
```

```python
# 3. 날짜 형식 통일 (예: YYYY-MM-DD)
df_main['TIME'] = pd.to_datetime(df_main['TIME']).dt.date
df_ext['날짜'] = pd.to_datetime(df_ext['날짜']).dt.date
```

```python
# 5. 필요한 칼럼만 선택
df_ext_avg = df_ext[['날짜', '평균']]
```

```python
# 6. 날짜 기준 병합 (left join 기준이면 기존 데이터 기준 유지)
df_merged = pd.merge(df_main, df_ext_avg, left_on='TIME', right_on='날짜', how='left')
```

```python
# 7. 필요 시 '날짜' 컬럼 제거
df_merged.drop(columns=['날짜'], inplace=True)
```

```python
df_merged.rename(columns={'평균': '한국 10년 채권수익률'}, inplace=True)
```

```python
# 결과 확인
print(df_merged.head())
```

#### **미국 채권 수익률 불러오기**

```python
df_ext2 = pd.read_csv("/content/미국 10년 채권수익률.csv")
df_ext2.head()
```

```python
# 3. 날짜 형식 통일 (예: YYYY-MM-DD)
df_main['TIME'] = pd.to_datetime(df_main['TIME']).dt.date
df_ext2['날짜'] = pd.to_datetime(df_ext2['날짜']).dt.date
```

```python
# 4. 종가, 시가, 고가, 저가 평균 계산
df_ext2['평균'] = df_ext2[['종가', '시가', '고가', '저가']].mean(axis=1)
```

```python
# 5. 필요한 칼럼만 선택
df_ext_avg2 = df_ext2[['날짜', '평균']]
```

#### **데이터 병합**

```python
# 6. 날짜 기준 병합 (left join 기준이면 기존 데이터 기준 유지)
df_merged2 = pd.merge(df_merged, df_ext_avg2, left_on='TIME', right_on='날짜', how='left')
```

```python
# 7. 필요 시 '날짜' 컬럼 제거
df_merged2.drop(columns=['날짜'], inplace=True)
```

```python
df_merged2.rename(columns={'평균': '미국 10년 채권수익률'}, inplace=True)
```

```python
# 결과 확인
print(df_merged2.head())
```
